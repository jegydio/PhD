# **Setup Python**


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

A linha `from PIL import Image` é uma instrução de importação em Python que carrega a classe `Image` do módulo PIL, que é um acrônimo para Python Imaging Library, mais recentemente conhecida como Pillow. Pillow é uma biblioteca de manipulação de imagens que adiciona suporte para abrir, manipular e salvar muitos formatos de arquivos de imagem diferentes.

```python
# Função para visualização das imagens
def exibir_imagens_lado_a_lado(imagem_original, imagem_filtrada,
                               titulo_original='Original',
                               titulo_filtrada='Filtrada',
                               tamanho_figura=(16, 8)):
    """
    Exibe duas imagens lado a lado para comparação.

    :param imagem_original: Array da imagem original.
    :param imagem_filtrada: Array da imagem filtrada.
    :param titulo_original: Título a ser exibido acima da imagem original.
    :param titulo_filtrada: Título a ser exibido acima da imagem filtrada.
    :param tamanho_figura: Tamanho da figura (largura, altura) em polegadas.
    """
    # Configurar o tamanho da figura para ter espaço suficiente para ambas as imagens
    plt.figure(figsize=tamanho_figura)

    # Primeira subfigura para a imagem original
    plt.subplot(1, 2, 1)  # 1 linha, 2 colunas, posição 1
    plt.imshow(imagem_original, cmap='gray')
    plt.title(titulo_original)  # Adiciona um título
    plt.axis('off')  # Para não mostrar os eixos

    # Segunda subfigura para a imagem filtrada
    plt.subplot(1, 2, 2)  # 1 linha, 2 colunas, posição 2
    plt.imshow(imagem_filtrada, cmap='gray')
    plt.title(titulo_filtrada)  # Adiciona um título
    plt.axis('off')  # Para não mostrar os eixos

    # Mostrar as imagens
    plt.show()
```

```python
# Visualizando a imagem fever-1
plt.imshow(image_array, cmap='gray')
plt.axis('off')  # Para não mostrar os eixos
plt.show()
```

#### Funções tratamento de ruído

```python
# Aplicando o filtro mediana utilizando o OpenCV
median_filtered_image = cv2.medianBlur(image_array, 3)  # Usando um kernel size igual a 3 (janela 3x3)
```

Este código aplica o filtro mediana a uma imagem usando a biblioteca OpenCV (Open Source Computer Vision Library), que é uma biblioteca utilizada para processamento de imagens e visão computacional em Python.

- `median_filtered_image = cv2.medianBlur(image_array, 3)`: Esta linha de código está realizando a operação de filtro mediana.

  - `cv2.medianBlur`: Esta função é uma operação de filtro mediana fornecida pelo OpenCV. O filtro mediana é um filtro de suavização não linear utilizado para reduzir o ruído em imagens. Ele funciona substituindo o valor de um pixel pela mediana dos valores de intensidade dos pixels no entorno desse pixel, dentro do tamanho de uma janela (ou kernel) especificado.

  - `image_array`: É a imagem de entrada sobre a qual o filtro será aplicado. Neste caso, espera-se que `image_array` seja uma matriz Numpy que representa uma imagem, possivelmente já convertida para escala de cinza como indicado nas etapas anteriores do processo.

  - `3`: Este é o tamanho do kernel, ou seja, a dimensão da janela de vizinhança utilizada para calcular a mediana. Um tamanho de kernel `3` significa que a janela tem dimensões 3x3 pixels. O tamanho do kernel influencia a quantidade de suavização: um kernel maior resulta em uma imagem mais suavizada, mas também pode causar a perda de detalhes finos. O tamanho do kernel deve ser um número ímpar para que haja um pixel central cujo valor será substituído pela mediana.

### **Obtendo ox e oy**

```python
# Carregar a imagem em escala de cinza
circulo_img_path = '/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 02 ]/circulo.png'
circulo_img = cv2.imread(circulo_img_path, cv2.IMREAD_GRAYSCALE)
```

1. Define uma string, `circulo_img_path`, que contém o caminho do arquivo onde a imagem `circulo.png` está localizada. Neste caso, o caminho é especificado para um ambiente Google Colab, onde a imagem está armazenada no Google Drive do usuário, dentro de uma série de pastas.

2. `cv2.imread`: Esta função é utilizada para ler a imagem do caminho fornecido. Ela pertence à biblioteca OpenCV, uma biblioteca popular para processamento de imagens e visão computacional.

3. O primeiro argumento da função `cv2.imread` é a variável `circulo_img_path`, que é o caminho para a imagem que você deseja abrir.

4. O segundo argumento, `cv2.IMREAD_GRAYSCALE`, indica que a imagem deve ser lida em escala de cinza. Isso significa que a imagem será convertida para uma imagem de um canal, onde os valores de cada pixel representam a intensidade da luz, ignorando as informações de cor. Isso é geralmente feito quando você está trabalhando com operações que não necessitam de cor, como detecção de bordas, reconhecimento de padrões em texto ou códigos de barras, etc.

5. `circulo_img`: A imagem lida pela função `cv2.imread` é então atribuída à variável `circulo_img`. Esta variável agora contém os dados da imagem em escala de cinza e pode ser usada para processamento adicional, como aplicação de filtros, detecção de bordas, etc.

```python
# Aplicar o filtro de Sobel no eixo x e no eixo y
sobelx = cv2.Sobel(circulo_img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(circulo_img, cv2.CV_64F, 0, 1, ksize=3)
```

- `sobelx = cv2.Sobel(circulo_img, cv2.CV_64F, 1, 0, ksize=3)`: Esta linha aplica o filtro de Sobel na direção horizontal (eixo x). Os parâmetros significam:
  - `circulo_img`: A imagem de entrada em escala de cinza.
  - `cv2.CV_64F`: O tipo de dados de saída para a imagem. `CV_64F` indica um float de 64 bits por pixel. É usado aqui para evitar possíveis problemas de subestimação durante a transformação, pois a detecção de bordas pode incluir mudanças negativas de intensidade.
  - `1, 0`: Estes são os parâmetros que indicam a direção do gradiente. `1` no eixo x e `0` no eixo y significa que estamos interessados na mudança de intensidade na direção x.
  - `ksize=3`: O tamanho do kernel utilizado pelo filtro de Sobel. Um kernel de tamanho 3x3 é comum e serve para calcular a derivada da imagem em uma pequena vizinhança de cada pixel.

- `sobely = cv2.Sobel(circulo_img, cv2.CV_64F, 0, 1, ksize=3)`: Esta linha faz praticamente o mesmo que a primeira, mas na direção vertical (eixo y). Os parâmetros `0, 1` significam que o filtro vai calcular a mudança de intensidade na direção y.

```python
# Converter as matrizes de volta para uint8
sobelx_uint8 = cv2.convertScaleAbs(sobelx)
sobely_uint8 = cv2.convertScaleAbs(sobely)
```

- `cv2.convertScaleAbs(sobelx)`: Esta função converte os valores de entrada na matriz `sobelx` para valores absolutos e depois os converte para `uint8`. Ao calcular o gradiente, os valores podem ser negativos ou positivos, e o tipo `CV_64F` pode ter valores fora do alcance de `uint8`. A conversão para valores absolutos garante que todos os valores negativos sejam mapeados para positivos de forma que possam ser representados em uma imagem em tons de cinza. Depois disso, a conversão para `uint8` garante que os valores sejam limitados ao intervalo de 0 a 255, adequados para a visualização da imagem.

- `cv2.convertScaleAbs(sobely)`: Faz o mesmo que a função anterior, mas para a matriz `sobely`, que contém o gradiente no eixo y.

```python
# visualizando ox e oy
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(sobelx_uint8, cmap='gray')
axes[0].set_title('Componente Ox')
axes[0].axis('off')
axes[1].imshow(sobely_uint8, cmap='gray')
axes[1].set_title('Componente Oy')
axes[1].axis('off')

plt.show()
```

```python
# Recarregando a imagem em escala de cinza e aplicando o filtro de Sobel
circulo_img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 02 ]/circulo.png',
                         cv2.IMREAD_GRAYSCALE)
sobelx = cv2.Sobel(circulo_img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(circulo_img, cv2.CV_64F, 0, 1, ksize=3)
```

```python
# Cálculo do módulo do gradiente
gradient_magnitude = cv2.magnitude(sobelx, sobely)
```

```python
# Gerando imagens para cada uma das operações
ver_todo = cv2.convertScaleAbs(sobelx)
ver_esq = cv2.convertScaleAbs(np.clip(sobelx, -np.inf, 0))
ver_dir = cv2.convertScaleAbs(np.clip(sobelx, 0, np.inf))
hor_todo = cv2.convertScaleAbs(sobely)
hor_sup = cv2.convertScaleAbs(np.clip(sobely, -np.inf, 0))
hor_inf = cv2.convertScaleAbs(np.clip(sobely, 0, np.inf))
modulo = cv2.convertScaleAbs(gradient_magnitude)
```

1. `ver_todo`: Esta linha cria uma imagem de todas as bordas verticais. Como `sobelx` contém as diferenças de intensidade no eixo x, convertendo para o tipo `uint8` com `cv2.convertScaleAbs`, você obtém uma imagem que destaca todas as bordas verticais.

2. `ver_esq`: Aqui, a função `np.clip` é usada para definir um limite inferior e superior nos valores de `sobelx`. Ao definir o limite superior como `0`, todos os valores positivos são recortados para `0`, mantendo apenas os valores negativos (bordas verticais esquerdas). Esses valores negativos são então convertidos para valores absolutos e para `uint8`.

3. `ver_dir`: Semelhante à linha acima, mas aqui os valores são recortados para `0` no lado inferior, mantendo apenas os valores positivos (bordas verticais direitas). Isso é então convertido para valores absolutos e para `uint8`.

4. `hor_todo`: Cria uma imagem de todas as bordas horizontais a partir do gradiente `sobely`, convertendo para `uint8`.

5. `hor_sup`: Usa `np.clip` para recortar os valores de `sobely` acima de `0`, mantendo os valores negativos que representam as bordas horizontais superiores, convertendo-os para valores absolutos e `uint8`.

6. `hor_inf`: O oposto de `hor_sup`, mantém valores positivos de `sobely`, recortando os negativos, para destacar as bordas horizontais inferiores.

7. `modulo`: Calcula o módulo do gradiente, que é a magnitude do vetor gradiente em cada ponto da imagem, combinando tanto `sobelx` quanto `sobely`. Isto dá uma medida da "força" da borda em cada ponto da imagem, independentemente da direção. A magnitude é convertida para `uint8` para facilitar a visualização.


```python
# Visualizando todas as imagens resultantes
images = [ver_todo, ver_esq, ver_dir, hor_todo, hor_sup, hor_inf, modulo]
titles = ['ver_todo', 'ver_esq', 'ver_dir', 'hor_todo', 'hor_sup', 'hor_inf', 'modulo']

# Mostrando as imagens com matplotlib
plt.figure(figsize=(15, 8))
for i in range(7):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
