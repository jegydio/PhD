## Transformando rotina ***Correção de perspectiva*** para Python

```python
# Define os pontos fonte e destino
src_points = np.array([[73, 0], [533, 0], [-22, 479], [629, 479]], dtype="float32")
dst_points = np.array([[16, 0], [630, 0], [14, 479], [630, 479]], dtype="float32")

# Calcula a matriz de transformação de perspectiva
m = cv2.getPerspectiveTransform(src_points, dst_points)

# Exibe a matriz de transformação
print("Matriz de Transformação de Perspectiva:")
print(m)

# Verifica a transformação com um ponto específico
v = np.array([[-22, 479, 1]], dtype="float64").T
w = np.dot(m, v)

# Corrige a perspectiva para exibir o ponto transformado
print("Ponto Transformado:")
print(w[0]/w[2], w[1]/w[2])

# Lê a imagem original
a = cv2.imread('/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/ka0.jpg')

# Aplica a transformação de perspectiva
b = cv2.warpPerspective(a, m, (a.shape[1], a.shape[0]))

# Salva a imagem transformada
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/ka11.jpg', b)

# Refaz a perspectiva invertendo a matriz de transformação
m_inv = np.linalg.inv(m)
a_inv = cv2.warpPerspective(b, m_inv, (b.shape[1], b.shape[0]))

# Salva a imagem com a perspectiva corrigida
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/ka22.jpg', a_inv)
```

```python
import numpy as np
from sklearn import neighbors
import matplotlib.pyplot as plt

def le(nomearq, colunas=None):
    with open(nomearq, 'r') as f:
        linhas = f.readlines()
    linha0 = linhas[0].split()
    n = len(linhas) - 1  # número de linhas de dados
    nc = int(linha0[1])  # número total de colunas nos dados
    # Lê somente o número de colunas especificado, se fornecido, senão lê todas as colunas
    colunas = colunas if colunas is not None else nc
    a = np.empty((n, colunas), dtype=np.float32)
    for l in range(n):
        linha = linhas[l+1].split()
        for c in range(colunas):
            a[l, c] = np.float32(linha[c])
    return a

##### main
ax = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/ax.txt", 4)
ay = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/ay.txt", 1)
qx = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/qx.txt", 4)
qy = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/qy.txt", 1)

vizinho = neighbors.KNeighborsClassifier(n_neighbors=1, weights='uniform', algorithm='brute')
vizinho.fit(ax, ay.ravel())

qp = vizinho.predict(qx)

erros = 0
for i in range(len(qp)):
    if qp[i] != qy[i]:
        erros += 1
print(f"Erros={erros}/{len(qp)}. Pct={100.0 * erros / len(qp):1.3f}%\n")

# Plota a superfície de decisão
fig = plt.figure(figsize=(12, 12), dpi=80)
for i in range(ax.shape[0]):
    if ay[i] == 1:
        plt.plot(ax[i, 0], ax[i, 1], 'ro')
    elif ay[i] == 2:
        plt.plot(ax[i, 0], ax[i, 1], 'go')
    elif ay[i] == 3:
        plt.plot(ax[i, 0], ax[i, 1], 'bo')
for i in range(qx.shape[0]):
    if qp[i] == 1:
        plt.plot(qx[i, 0], qx[i, 1], 'rx')
    elif qp[i] == 2:
        plt.plot(qx[i, 0], qx[i, 1], 'gx')
    elif qp[i] == 3:
        plt.plot(qx[i, 0], qx[i, 1], 'bx')

minx = min(ax[:, 0].min(), qx[:, 0].min()) - 1
miny = min(ax[:, 1].min(), qx[:, 1].min()) - 1
maxx = max(ax[:, 0].max(), qx[:, 0].max()) + 1
maxy = max(ax[:, 1].max(), qx[:, 1].max()) + 1

xx, yy = np.meshgrid(np.arange(minx, maxx, 0.05), np.arange(miny, maxy, 0.05))

# Cria a grade para a superfície de decisão incluindo colunas de zero para o "ruído"
r1, r2 = xx.flatten(), yy.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))

# Adiciona colunas de zeros para o "ruído"
zeros_col = np.zeros((len(r1), 2), dtype=np.float32)
grid = np.hstack((r1, r2, zeros_col))

# Faz a predição em toda a grade
yhat = vizinho.predict(grid)
zz = yhat.reshape(xx.shape)

# Plota a superfície de decisão
plt.contourf(xx, yy, zz, cmap='Paired')
plt.tight_layout()
plt.savefig("noysenote2.png")
plt.show()
```

```python
import numpy as np
from sklearn import neighbors
import matplotlib.pyplot as plt

def le(nomearq, colunas=None):
    with open(nomearq, 'r') as f:
        linhas = f.readlines()
    linha0 = linhas[0].split()
    n = len(linhas) - 1  # número de linhas de dados
    nc = int(linha0[1])  # número total de colunas nos dados
    # Lê somente o número de colunas especificado, se fornecido, senão lê todas as colunas
    colunas = colunas if colunas is not None else nc
    a = np.empty((n, colunas), dtype=np.float32)
    for l in range(n):
        linha = linhas[l+1].split()
        for c in range(colunas):
            a[l, c] = np.float32(linha[c])
    return a

##### main
ax = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/ax.txt", 4)
ay = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/ay.txt", 1)
qx = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/qx.txt", 4)
qy = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/qy.txt", 1)

vizinho = neighbors.KNeighborsClassifier(n_neighbors=1, weights='uniform', algorithm='kd_tree')
vizinho.fit(ax, ay.ravel())

qp = vizinho.predict(qx)

erros = np.sum(qp != qy.ravel())
print(f"Erros={erros}/{len(qp)}. Pct={100.0 * erros / len(qp):1.3f}%\n")

# Define os limites para a figura
minx, miny = ax[:, :2].min(axis=0) - 1
maxx, maxy = ax[:, :2].max(axis=0) + 1

# Aumenta o passo da malha para acelerar a plotagem
step_size = 0.1
xx, yy = np.meshgrid(np.arange(minx, maxx, step_size), np.arange(miny, maxy, step_size))

# Cria a grade para a superfície de decisão com dois atributos fictícios para as colunas de ruído
grid = np.c_[xx.ravel(), yy.ravel(), np.zeros((xx.ravel().shape[0], 2))]

# Faz a predição em toda a grade
yhat = vizinho.predict(grid)
zz = yhat.reshape(xx.shape)


# Plota a superfície de decisão
plt.figure(figsize=(12, 12))
cmap = plt.cm.coolwarm
plt.contourf(xx, yy, zz, alpha=0.8, cmap=cmap)

# Plota os pontos de treinamento com cores para cada classe
colors = ['red', 'green']  # Duas cores para duas classes
class_labels = ['Class 1', 'Class 2']
for i, color in enumerate(colors):
    class_mask = ay.ravel() == (i + 1)
    plt.scatter(ax[class_mask, 0], ax[class_mask, 1], c=color, label=class_labels[i], edgecolor='k')

plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Surface with KNN')
plt.tight_layout()
plt.show()
```

## Árvore de decisão

```python
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def le(nomearq, colunas=None):
    with open(nomearq, 'r') as f:
        linhas = f.readlines()
    linha0 = linhas[0].split()
    n = len(linhas) - 1
    nc = int(linha0[1])
    colunas = colunas if colunas is not None else nc
    a = np.empty((n, colunas), dtype=np.float32)
    for l in range(n):
        linha = linhas[l+1].split()
        for c in range(colunas):
            a[l, c] = np.float32(linha[c])
    return a

##### main
ax = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/ax.txt", 4)
ay = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/ay.txt", 1)
qx = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/qx.txt", 4)
qy = le("/content/drive/MyDrive/Colab Notebooks/_phd/[ 2024 - PSI5790 ]/[ Aula 04 ]/noisynote/qy.txt", 1)

arvore = tree.DecisionTreeClassifier()
arvore.fit(ax, ay.ravel())

qp = arvore.predict(qx)

erros = np.sum(qp != qy.ravel())
print(f"Erros={erros}/{len(qp)}. Pct={100.0 * erros / len(qp):1.3f}%\n")

# Define os limites da figura e cria a malha
minx, miny = ax[:, :2].min(axis=0) - 1
maxx, maxy = ax[:, :2].max(axis=0) + 1
step_size = 0.1
xx, yy = np.meshgrid(np.arange(minx, maxx, step_size), np.arange(miny, maxy, step_size))

# Prepara os dados para a superfície de decisão
grid = np.c_[xx.ravel(), yy.ravel(), np.zeros((xx.ravel().shape[0], 2))]

# Faz a predição em toda a grade
yhat = arvore.predict(grid)
zz = yhat.reshape(xx.shape)

# Plota a superfície de decisão
plt.figure(figsize=(10, 8))
cmap_light = plt.cm.coolwarm
plt.contourf(xx, yy, zz, alpha=0.3, cmap=cmap_light)

# Plota os pontos de treinamento
scatter = plt.scatter(ax[:, 0], ax[:, 1], c=ay.ravel(), cmap=cmap_light, edgecolor='k', label='Training points')

# Cria a legenda dos pontos
classes = np.unique(ay.ravel())
class_labels = ["Class " + str(int(cls)) for cls in classes]

# Obter cores da paleta de cores para cada classe
class_colours = [cmap_light(cls / max(classes)) for cls in classes]
recs = [mpatches.Rectangle((0, 0), 1, 1, fc=class_colours[i]) for i in range(len(class_colours))]
plt.legend(recs, class_labels, loc='upper right')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Surface of Decision Tree')
plt.tight_layout()
plt.show()
```
